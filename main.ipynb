```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread('./images/beyin.jpg', cv.IMREAD_GRAYSCALE)
```

```python
blurred = cv.GaussianBlur(image, (5,5), 0)

plt.imshow(blurred, cmap='gray'), plt.title('Gaussian Blur')
plt.show()
```

```python
clahe = cv.createCLAHE(clipLimit=2, tileGridSize=(8,8))
clahe_img = clahe.apply(image)

plt.imshow(clahe_img, cmap='gray'), plt.title('CLAHE')
plt.show()
```

```python
canny = cv.Canny(clahe_img, 125, 175)

plt.imshow(canny, cmap='gray'), plt.title('Canny')
plt.show()
```

```python
ret, thresh = cv.threshold(clahe_img, 175, 255, cv.THRESH_BINARY)

plt.imshow(thresh, cmap='gray'), plt.title('Threshold')
plt.show()
```
